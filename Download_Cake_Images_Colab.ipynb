````
# Download all images
downloaded = 0
failed = 0

for i, url in enumerate(CAKE_IMAGE_URLS):
    try:
        response = requests.get(url, timeout=30)
        if response.status_code == 200:
            # Open and resize image
            img = Image.open(BytesIO(response.content))
            img = img.convert('RGB')
            img = img.resize((512, 512))
            
            # Save
            filename = f"cake_{i+1:03d}.jpg"
            img.save(os.path.join(FOLDER, filename), quality=95)
            downloaded += 1
            print(f"‚úÖ Downloaded: {filename}")
        else:
            failed += 1
            print(f"‚ùå Failed: {url[:50]}...")
    except Exception as e:
        failed += 1
        print(f"‚ùå Error: {str(e)[:50]}")
    
    time.sleep(0.5)  # Be nice to servers

print(f"\nüìä Summary: {downloaded} downloaded, {failed} failed")
````

````
# Show downloaded images
import matplotlib.pyplot as plt

images = os.listdir(FOLDER)
print(f"\nüì∏ Total images in folder: {len(images)}")

# Display first 12 images
fig, axes = plt.subplots(3, 4, figsize=(15, 12))
for i, ax in enumerate(axes.flat):
    if i < len(images):
        img = Image.open(os.path.join(FOLDER, images[i]))
        ax.imshow(img)
        ax.set_title(images[i])
    ax.axis('off')
plt.tight_layout()
plt.show()
````
